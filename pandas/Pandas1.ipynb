```python
import pandas as pd

import numpy as np
```

```python
n1 = np.array([[2,1,4,3],[4,5,6,2],[3,5,4,6]])
print(n1)
```

```python
n2 = np.array([[2,1,4,3],[4,5,6,2],[3,5,4.1,6]])
print(n2)
```

```python
n3 = np.array([[2,1,4,3],[4,'5',6,2],[3,5,4.1,6]])
print(n3)
```

```python
d1 = pd.DataFrame(n1)
d1.columns = ['a','b','c','d']
d1.index = ['r1','r2','r3']
d1
```
