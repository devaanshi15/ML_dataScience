**Treating missing values in a data frame**

```python
df.isna().sum()
```

```python
print("Percentage of missing values - ")
print()
print((df.isna().sum()/len(df))*100)
```

```python
#Cabin column has almost 77% values are missing, which means this column is having majority of values missing
#it makes this column a less important column for this data
#So we can remove this data column

#drop() -> to drop a specific row or a column

df1 = df.drop(columns = ['Cabin'])
```

```python
df1.head(3)
```

```python
df1.isna().sum()
```

```python
#Embarked column has just 2 values missing - removing these two rows will not much affect the data

#dropna() - used to drop all the rows with missing values

df2 = df1.dropna(subset = ['Embarked'])
df2.head(3)
```

```python
df2.isna().sum()
```

```python
#Age => It can be an importnat data column with some important insights about data
#We have more then 80% values are known
#We can calculate mean or median to fill the blank spots

print(df2['Age'].mean())
print(df2['Age'].median())
```

```python
#fillna() => to fill the missing values existing in a data column

df2['Age'] = df2['Age'].fillna(29)
```

```python
df2.isna().sum()
```

```python
print(df2['Age'].mean())
print(df2['Age'].median())
```

**filling up missing values with leading or previous value in the cell**

```python
df_x1 = df.copy()
```

```python
df_x1.isna().sum()
```

```python
#find out where the values are missing in embarked
df_x1[df_x1['Embarked'].isna()]
```

```python
df_x1.loc[[60,61,62,828,829,830]]
```

```python
#fill Embarked missing values with previous cell value in the data column

df_x1['Embarked'] = df_x1['Embarked'].fillna(method = 'pad')
```

```python
df_x1.loc[[60,61,62,828,829,830]]
```

```python
df_x2 = df.copy()
df_x2.isna().sum()
```

```python
df_x2.loc[[60,61,62,828,829,830]]
```

```python
#fill Embarked missing values with next cell value in the data column

df_x2['Embarked'] = df_x2['Embarked'].fillna(method = 'bfill')
```
